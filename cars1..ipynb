```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
Auto=pd.read_csv("Automobile_data.csv")
Auto
```

```python
Auto.info() 
```

```python
Auto["normalized-losses"].replace("?",np.nan,inplace=True)
Auto["horsepower"].replace("?",np.nan,inplace=True)
Auto
```

```python
Auto.isnull().sum()
```

```python
Auto["normalized-losses"]=Auto["normalized-losses"].astype("float64")
Auto["horsepower"]=Auto["horsepower"].astype("float64")
Auto
```

```python
Auto.describe()
```

```python
Auto["normalized-losses"].fillna(122,inplace=True)
Auto["horsepower"].fillna(104,inplace=True)
Auto
```

```python
Auto.isnull().sum()
```

```python
Auto.info()
```

```python
Auto_num=Auto.select_dtypes(["float64","int64"])
Auto_num
```

```python
Auto_cat=Auto.select_dtypes(["object"])  
Auto_cat
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in Auto_cat:
    le=LabelEncoder()
    Auto_cat[col]=le.fit_transform(Auto_cat[col])
```

```python
Auto_cat
```

```python
Auto2=pd.concat([Auto_num,Auto_cat],axis=1)
```

```python
Auto2
```

```python
X=Auto2.drop("price",axis=1) 
Y=Auto["price"] 
```

```python
X
```

```python
Y
```

```python
#TRAIN TEST SPLIT

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
y_pred=lr.predict(X_train)
```

```python
y_pred
```

```python
y_predtest=lr.predict(X_test)
```

```python
y_predtest
```

```python
r2_score(Y_train,y_pred)
```

```python
r2_score(Y_test,y_predtest)
```

```python
lr.score(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
l2=Ridge(0.01) #ALPHA VALUE IN BRACKET
l2.fit(X_train,Y_train)
```

```python
l2.score(X_train,Y_train)
```

```python
l2.score(X_test,Y_test)
```

```python
l2.coef_
```

```python
#RIDGE

for alpha in range(1,5):
    l2=Ridge(alpha)
    l2.fit(X_train,Y_train)
    train_score=l2.score(X_train,Y_train)
    test_score=l2.score(X_test,Y_test)
    print("Alpha:",alpha)
    print("train score:",train_score)
    print("test score:",test_score)
    print("-----------------") #0.8504568715278975
    
    #0.7965544858354374
```

```python
#LASSO
l3=Lasso(1000)
l3.fit(X_train,Y_train)
```

```python
l3.score(X_train,Y_train)
```

```python
l3.score(X_test,Y_test)
```

```python
l3.coef_
```

```python
for i in range(200,500,50):
    l3=Lasso(i)
    l3.fit(X_train,Y_train)
    train_score=l3.score(X_train,Y_train)
    test_score=l3.score(X_test,Y_test)
    print("Alpha:",i)
    print("train score:",train_score)
    print("test score:",test_score)
    print("-----------------")
```

```python
l3.coef_
```

```python
#FINAL RIDGE MODEL==SELECTING ALPHA 2

l2=Ridge(alpha=2)
l2.fit(X_train,Y_train)
print(l2.score(X_test,Y_test))
```

```python
l2.coef_
```

```python
#FINAL LASSO MODEL==SELECTING ALPHA 200

l3=Lasso(alpha=200)
l3.fit(X_train,Y_train)
print(l3.score(X_test,Y_test))
```

```python
l3.coef_
```

```python
#CROSS VALIDATION

from sklearn.model_selection import cross_val_score
```

```python
l2_cross=cross_val_score(l2,X,Y,cv=4) #cv should be always between 3 to 5
```

```python
l2_cross
```

```python
np.mean(l2_cross) # if the mean is greater 0.5 then the algorithm is working properly
```

```python
l3_cross=cross_val_score(l3,X,Y,cv=4)
```
